```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
import warnings 
warnings.filterwarnings( "ignore" ) 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv( "country_vaccinations.csv" ) 
df.head(20)
```

```python
df1=df.drop(['people_vaccinated','people_fully_vaccinated','daily_vaccinations_raw','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred'],axis=1)
df1.head(20)
```

```python
mean_value=df1['total_vaccinations'].mean()
print(mean_value)
```

```python
df1['total_vaccinations'].fillna(value=df1['total_vaccinations'].mean(), inplace=True)
```

```python
df1.head(20)
```

```python
df1['daily_vaccinations'].fillna(value=df1['daily_vaccinations'].mean(), inplace=True)
df1.head()
```

```python
df1['total_vaccinations_per_hundred'].fillna(value=df1['total_vaccinations_per_hundred'].mean(), inplace=True)
df1.head()
```

```python
df1['daily_vaccinations_per_million'].fillna(value=df1['daily_vaccinations_per_million'].mean(), inplace=True)
df1.head()
```

```python
df1.info()
```

```python
cols_to_norm = ['total_vaccinations','daily_vaccinations','total_vaccinations_per_hundred','daily_vaccinations_per_million']
df1[cols_to_norm] = df1[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df1.head()
```

```python
df1.plot.scatter(x = 'total_vaccinations', y = 'total_vaccinations_per_hundred')
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= df1['total_vaccinations']
y=df1['total_vaccinations_per_hundred']
z=df1['daily_vaccinations_per_million']
# Creating figure
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
   
# Add x, y gridlines 
ax.grid(b = True, color ='grey', 
        linestyle ='-.', linewidth = 0.3, 
        alpha = 0.2) 
 
 
# Creating color map
my_cmap = plt.get_cmap('hsv')
 
# Creating plot
sctt = ax.scatter3D(x,y,z,alpha = 0.8,
                    c = (x + y + z), 
                    cmap = my_cmap, 
                    marker ='^')

plt.title("simple 3D scatter plot")
ax.set_xlabel('total_vaccinations', fontweight ='bold') 
ax.set_ylabel('total_vaccinations_per_hundred', fontweight ='bold') 
ax.set_zlabel('daily_vaccinations_per_million', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
 
# show plot
plt.show()
```
